```python
import torch as t
```

### Example 1:-

```python
#Intialization
x = t.rand(1,requires_grad=True)
print(x)
y = x**2
print(y)
```

```python
# dx/dy 
y.backward()
x.grad
```

```python

def dy_dx(x):
    return 2*x*t.cos(x**2)
x = t.rand(1, requires_grad=True)
z = dy_dx(x)
```

```python
z.backward()
x.grad
```
